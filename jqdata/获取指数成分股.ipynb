```python
from jqdata import *
import pandas as pd
from datetime import datetime,timedelta
```

```python
def code_convert(code):
    if 'XSHE' in code:
        code = code.replace("XSHE",'SZ')
    else:
        code = code.replace("XSHG",'SH')
        
    return code
```

```python
today = datetime.now()
end = today#.strftime('%Y-%m-%d')
start = '2005-01-01'

days = get_trade_days(start_date = start, end_date = end)
```

```python
code_list = ['000852.XSHG','000905.XSHG','000300.XSHG','000016.XSHG']
```

# 获取历史指数成分股

```python
for index_code in code_list:
    data = []
    pre_codes = []
    for day in days:
        date = day.strftime('%Y-%m-%d')
        codes = get_index_stocks(index_code, date=date)

        codes = list(map(lambda x:code_convert(x),codes))

        codes.sort()
        if pre_codes == codes:
            continue

        pre_codes = codes

        data.append({
            'date':date,
            'codes':codes
        })
    df = pd.DataFrame(data)
    
    df = df.set_index('date')
    
    file = code_convert(index_code)
    df.to_csv(f"{file}.csv")
```
