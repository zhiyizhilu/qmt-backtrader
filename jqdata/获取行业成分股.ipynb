```python
def code_convert(code):
    if 'XSHE' in code:
        code = code.replace("XSHE",'SZ')
    else:
        code = code.replace("XSHG",'SH')
        
    return code
```

# 获取历史行业成分股

```python
for industry_code in INDUSTRIES.keys():
    print(f"正在获取:{industry_code}")
    
    data = []
    pre_codes = []
    for day in days:
        date = day.strftime('%Y-%m-%d')
        codes = get_industry_stocks(industry_code, date=date)

        codes = list(map(lambda x:code_convert(x),codes))

        codes.sort()
        if pre_codes == codes:
            continue

        pre_codes = codes

        data.append({
            'date':date,
            'codes':codes
        })
        
    df = pd.DataFrame(data)
    df = df.set_index('date')
    
    
    file = INDUSTRIES[industry_code]
    df.to_csv(f"{file}.csv")
```
